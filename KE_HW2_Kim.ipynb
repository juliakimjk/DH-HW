```python
import re
import natasha
import os
from natasha import (
    Segmenter,
    MorphVocab,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    PER,
    NamesExtractor,

    Doc
)
```

```python
segmenter = Segmenter()
morph_vocab = MorphVocab()
emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)
names_extractor = NamesExtractor(morph_vocab)
```

```python
texts = []
for i in os.listdir('D:/corp/raw_corpus/Lenta.ru/lenta2010/')[:10]:
    with open('D:/corp/raw_corpus/Lenta.ru/lenta2010/' + i, 'r', encoding='UTF-8') as f:
        lines = f.readlines()
        for line in lines[2:]:
            texts.append(line)
```

```python
for text in texts:
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.tag_ner(ner_tagger)
    doc.ner.print()
    persons = []
    organizations = []
    locations = []
    for span in doc.spans:
        if span.type == 'PER':
            persons.append(span.text)
        elif span.type == 'ORG':
            organizations.append(span.text)
        elif span.type == 'LOC':
            locations.append(span.text)
    print('Organizations - ', len(organizations))
    print('Persons - ', len(persons))
    print('Locations - ', len(locations))
```
